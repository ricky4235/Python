```python
def GetGoods(response):
    soup = BeautifulSoup(response.text)
    url = soup.find('link')['href']
    ndf = pd.DataFrame(data = [{'name':soup.select_one("h2.goodsName").text.strip(),
                                'price':soup.select_one("td.priceArea").b.text.strip(),
                                'brand':soup.select_one("a.brandNameTxt").text.strip(),
                                'site':soup.select_one("a.openClassification classification")["org"] }],
                        good = ['name', 'price', 'brand', 'site'] )
    return ndf
```

```python
def GetLinks(response):
    links = []
    soup = BeautifulSoup(response.text)
    for i in soup.find_all('div', 'product-gallery-item '):
        url = i.find('a')['href']
        links.append(url)
    return links
links
```
