ç¢ºèªå¯ä»¥é€éç¨‹å¼æŠŠæ–‡ç« çˆ¬ä¸‹ä¾†ä¹‹å¾Œï¼Œæˆ‘å€‘å°±ä¾†å¯«å€‹ç°¡å–®çš„Crawlå‡½æ•¸ï¼ŒæœŸæœ›åªéœ€è¦è¼¸å…¥æ–‡ç« çš„IDå¾Œï¼Œå°±å›å‚³çˆ¬ä¸‹ä¾†çš„æ–‡ç« å…§å®¹ï¼

```python
# æ’°å¯«ç°¡å–®çš„å‡½æ•¸ï¼Œé€éè¼¸å…¥æ–‡ç« IDï¼Œå°±è¼¸å‡ºæ–‡ç« çš„è³‡æ–™
def Crawl(ID):
    link = 'https://www.dcard.tw/_api/posts/' + str(ID)
    requ = requests.get(link)
    rejs = requ.json()
    return(pd.DataFrame(
        data=
        [{'ID':rejs['id'],
          'title':rejs['title'],
          'content':rejs['content'],
          'excerpt':rejs['excerpt'],
          'createdAt':rejs['createdAt'],
          'updatedAt':rejs['updatedAt'],
          'commentCount':rejs['commentCount'],
          'forumName':rejs['forumName'],
          'forumAlias':rejs['forumAlias'],
          'gender':rejs['gender'],
          'likeCount':rejs['likeCount'],
          'reactions':rejs['reactions'],
          'topics':rejs['topics']}],
        columns=['ID','title','content','excerpt','createdAt','updatedAt','commentCount','forumName','forumAlias','gender','likeCount','reactions','topics']))
```

```python
# å˜—è©¦ä½¿ç”¨æ’°å¯«å‡ºçš„å‡½æ•¸ï¼ŒæŠ“å–ç·¨è™Ÿ231030181çš„æ–‡ç« 
Crawl(232646236)
```

# ä¸€æ¬¡è®€å–100ç¯‡æœ€ç†±é–€çš„æ–‡ç« 
é€éç³»çµ±æä¾›çš„apiï¼Œä¸€æ¬¡æŠ“å–100ç¯‡ç†±é–€æ–‡ç« <br>
https://www.dcard.tw/_api/posts?popular=true&limit=100

ä»¥ä¸‹ç°¡å–®èªªæ˜é€™å€‹ç¶²å€

* popularåƒæ•¸ï¼šè‹¥è¨­å®šç‚ºtrueï¼Œè¡¨ç¤ºæŒ‰ç…§ç†±é–€ç¨‹åº¦æ’åºï¼Œè‹¥è¨­å®šç‚ºfalseï¼Œå‰‡æŒ‰ç…§ç™¼å¸ƒæ™‚é–“æ’åº
* limitåƒæ•¸ï¼šé™å®šåœ¨0-100çš„æ•¸å€¼ï¼Œè¡¨ç¤ºè¦æŠ“å¤šå°‘æ–‡ç« 

```python
url = 'https://www.dcard.tw/_api/posts?popular=true&limit=100'
resq = requests.get(url)
rejs = resq.json()
df = pd.DataFrame()
for i in range(len(rejs)):
    df = df.append(Crawl(rejs[i]['id']),ignore_index=True)
print(df.shape)
df
```

```python
# çˆ¬è¶…é100ç¯‡Dcardæ–‡ç« 
ç”±æ–¼APIé™åˆ¶æœ€å¤šè¼‰å…¥100ç¯‡æ–‡ç« ï¼Œå¦‚æœæˆ‘å€‘æƒ³è¦çˆ¬æ›´å¤šè³‡æ–™ï¼Œå¯ä»¥é€ébeforeåƒæ•¸èˆ‡è¿´åœˆé€²è¡Œ!<br>
beforeåƒæ•¸å¾Œé¢æ˜¯æ¥æ–‡ç« çš„IDï¼Œè®“æˆ‘å€‘å¯ä»¥æŠ“å–æŸç¯‡æ–‡ç« ä¹‹å‰çš„è³‡æ–™<br>
è€Œé€éè¿´åœˆï¼Œæˆ‘å€‘åªéœ€è¦æŠŠä¹‹å‰æŠ“åˆ°æœ€å¾Œä¸€ç¯‡æ–‡ç« çš„IDæ”¾å…¥beforeåƒæ•¸ä¸­ï¼Œæˆ‘å€‘å°±å¯ä»¥æŠ“åˆ°é€™ç¯‡æ–‡ç« çš„å‰100ç¯‡æ–‡ç« ã€‚
```

```python
# é€éè¿´åœˆè®€å–10*100ç¯‡æ–‡ç« ï¼Œè‹¥éœ€è®€å–æ›´å¤šè³‡æ–™ï¼Œå¯ä»¥å°‡range(10)ä¸­çš„æ•¸å€¼æå‡
for j in range(2):
    last = str(int(df.tail(1).ID)) # æ‰¾å‡ºçˆ¬å‡ºè³‡æ–™çš„æœ€å¾Œä¸€ç­†IDï¼Œtail()å’Œhead()ä¸€æ¨£ï¼Œå¯ä»¥åœ¨æ‹¬è™Ÿå…§åŠ ä¸Šè¦é¡¯ç¤ºå¹¾ç­†è³‡æ–™
    url = 'https://www.dcard.tw/_api/posts?popular=true&limit=100&before=' + last
    resq = requests.get(url)
    rejs = resq.json()
    for i in range(len(rejs)):
        df = df.append(Crawl(rejs[i]['id']), ignore_index=True)
print(df.shape)
df
```
