```python
import pandas as pd
import re, time, requests
from selenium import webdriver
from bs4 import BeautifulSoup
```

# æœå°‹è²¼æ–‡é€£çµ

### ç¢°åˆ°å•é¡Œï¼š
1. ç€‘å¸ƒå¼ç¶²é ï¼šç¶²é ä¸æœƒä¸€æ¬¡æŠŠæ‰€æœ‰è²¼æ–‡éƒ½åŠ è¼‰çµ¦ä½ ï¼Œç•¶æˆ‘å€‘æŠŠé é¢æ»¾å‹•åˆ°æœ€ä¸‹ç«¯æ™‚æ‰æœƒåŠ è¼‰æ–°è²¼æ–‡ã€‚
    * è§£æ±ºæ–¹æ¡ˆï¼šé€éjavaæŒ‡ä»¤è®“è¦–çª—æ»¾å‹•åˆ°åº•éƒ¨
2. ç³»çµ±å½ˆçª—ï¼šç•¶æˆ‘å€‘æŠŠé é¢å¾€ä¸‹æ»¾å‹•å¾Œï¼Œæœƒå½ˆå‡ºä¸€å€‹è«‹æˆ‘å€‘ç™»å…¥æˆ–è¨»å†Šçš„è¦–çª—ï¼Œé˜»ç¤™æˆ‘å€‘çˆ¬è³‡æ–™çš„æµç¨‹ã€‚
    * è§£æ±ºæ–¹æ¡ˆï¼šåµæ¸¬ã€ŒNot Nowã€çš„elementä½ç½®ï¼Œä¸¦é€éç¨‹å¼é»æ“Šé€™å€‹element

```python
def FindLinks(url, n):
    Links = []
    driver.get(url)
    for i in range(n):
        time.sleep(2)
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);') #æ»¾å‹•ç¶²é åˆ°åº•éƒ¨çš„å‘½ä»¤
    driver.find_element_by_xpath('//a[@id="expanding_cta_close_button"]').click() # é€™è£¡æœƒè·³å‡ºè¦æ±‚ç™»å…¥çš„ç•«é¢ï¼Œæ‰¾åˆ°ã€Œç¨å¾Œå†èªªã€çš„æŒ‰éˆ•ä¸¦é»æ“Š
    soup = BeautifulSoup(driver.page_source)
    posts = soup.findAll('div', {'class':'clearfix y_c3pyo2ta3'})
    for i in posts:  #éæ­·æ¯å€‹è²¼æ–‡çš„é€£çµ
        Links.append('https://www.facebook.com' + i.find('a',{'class':'_5pcq'}).attrs['href'].split('?',2)[0]) 
    return Links
```

### å±•é–‹æ‰€æœ‰ç•™è¨€
å®šç¾©ä¸€å€‹å±•é–‹æ‰€æœ‰ç•™è¨€çš„å‡½æ•¸ï¼Œé€éwhileè¿´åœˆåè¦†æœå°‹èˆ‡é»æ“Šã€Œçœ‹æ›´å¤šç•™è¨€ã€ã€ã€Œçœ‹æ›´å¤šå›è¦†ã€èˆ‡ã€Œçœ‹å®Œæ•´è²¼æ–‡å…§å®¹ã€ç­‰æŒ‰éˆ•ã€‚
åœ¨éç¨‹ä¸­æœƒå‡ºç¾è«‹æˆ‘å€‘ç™»å…¥æˆ–è¨»å†Šçš„å½ˆè·³è¦–çª—ï¼Œä½†æˆ‘å€‘ä¸ç¢ºå®šåˆ°åº•ä»€éº¼æ™‚å€™æœƒè·³å‡ºï¼Œå› æ­¤éœ€è¦åœ¨éç¨‹ä¸­åè¦†åµæ¸¬æ˜¯æœ‰å‡ºç¾é€™å€‹å½ˆçª—ï¼Œè‹¥æœ‰å°±é»æ“Šã€ŒNot Nowã€

```python
def expand(url):
    driver.get(url)
    try:
        driver.find_element_by_xpath('//a[@lang="en_US"]').click()
    except:
        print("Now is in EN_US")
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);') #éç¨‹ä¸­éš¨æ™‚å¯èƒ½æœƒå‡ºç¾å½ˆçª—ï¼Œè‹¥æœ‰å°±é»æ“Šã€ŒNot Nowã€
    
    # é»æ“Šã€Œcommentsã€ï¼Œè—‰ä»¥å±•é–‹ç•™è¨€
    try:
        driver.find_element_by_xpath('//div[@class="_5pcr userContentWrapper"]//a[@data-testid="UFI2CommentsCount/root"]').click()
        time.sleep(1)
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);') #éç¨‹ä¸­éš¨æ™‚å¯èƒ½æœƒå‡ºç¾å½ˆçª—ï¼Œè‹¥æœ‰å°±é»æ“Šã€ŒNot Nowã€
        time.sleep(1)
        driver.find_element_by_id('expanding_cta_close_button').click() 
    except:
        print('There is no comment!')
    k = 1
    while k != 0:
        k = 0
        for i in driver.find_elements_by_xpath('//div[@class="_5pcr userContentWrapper"]//div[@data-testid="UFI2CommentsList/root_depth_0"]//a[@role="button"]'): 
            # åè¦†åµæ¸¬æ˜¯å¦æœ‰ã€Œçœ‹æ›´å¤šç•™è¨€ã€ã€ã€Œçœ‹æ›´å¤šå›è¦†ã€èˆ‡ã€Œçœ‹å®Œæ•´è²¼æ–‡å…§å®¹ã€ç­‰æŒ‰éˆ•ï¼Œè‹¥æœ‰æ“‡é»æ“Š
            if bool(re.search('comment|More|Repl',i.text)) == True :
                driver.execute_script('window.scrollTo(0, document.body.scrollHeight);') #éç¨‹ä¸­éš¨æ™‚å¯èƒ½æœƒå‡ºç¾å½ˆçª—ï¼Œè‹¥æœ‰å°±é»æ“Šã€ŒNot Nowã€
                time.sleep(2)
                try:
                    driver.find_element_by_xpath('//div[@style="display: block;"]//a[@id="expanding_cta_close_button"]').click()
                except:
                    print('No pupup!')
                try:
                    i.click()
                except:
                    print('Nothing')
                time.sleep(2)
                k += 1
```

### æ“·å–è²¼æ–‡è³‡è¨Šèˆ‡äº’å‹•æ‘˜è¦
é€éé€å±¤æœç´¢çš„æ–¹å¼ï¼Œé€æ­¥å®šä½æˆ‘å€‘è¦æ‰¾çš„è³‡è¨Š
åœ¨é€™å€‹ç’°ç¯€éœ€è¦åè¦†é€éChromeçš„åŠŸèƒ½æ¯”å°è³‡æ–™ï¼Œéœ€è¦èŠ±ä¸€äº›å¿ƒåŠ›é€²è¡Œæ¯”å°
å¦å¤–åœ¨é€™éƒ¨åˆ†ä¹Ÿä½¿ç”¨äº†å¤§é‡å€‹try-exceptï¼ŒåŸå› æ˜¯è¨±å¤šè³‡æ–™æ˜¯æœ‰å…§å®¹æ‰æœƒå‡ºç¾ã€‚ä¾‹å¦‚ä¸¦éæ¯å¤©è²¼äººéƒ½æœƒæ”¶åˆ°ã€Œå“ˆå“ˆã€ã€ã€Œç”Ÿæ°£ã€çš„å¿ƒæƒ…ã€‚

```python
# æ–‡ç« å…§å®¹èˆ‡äº’å‹•æ‘˜è¦
def PostContent(soup):
    # poæ–‡å€å¡Š
    userContent = soup.find('div', {'class':'_5pcr userContentWrapper'})
    # poæ–‡äººè³‡è¨Šå€å¡Š
    PosterInfo = userContent.find('div', {'class':'l_c3pyo2v0u i_c3pynyi2f clearfix'})
    # äº’å‹•æ‘˜è¦å€(è®šã€ç•™è¨€èˆ‡åˆ†äº«)
    feedback = soup.find('form', {'class':'commentable_item collapsed_comments'})
    # åç¨±
    Name = PosterInfo.find('img').attrs['aria-label']
    # ID
    ID = PosterInfo.find('a', {'class':'_5pb8 o_c3pynyi2g _8o _8s lfloat _ohe'}).attrs['href'].split('/?',2)[0].split('/',-1)[-1]
    # ç¶²å€
    Link = driver.current_url
    # ç™¼æ–‡æ™‚é–“
    try:
        Time = PosterInfo.find('abbr').attrs['title']
    except:
        Time = PosterInfo.find('div', {'class':'_1atc fsm fwn fcg'}).text
    # æ–‡ç« å…§å®¹
    try:
        Content = userContent.find('div', {'class':'_5pbx userContent _3576'}).text
    except:
        Content = ""
    # Like
    try:
        Like = feedback.find('span', {'data-testid':'UFI2TopReactions/tooltip_LIKE'}).find('a').attrs['aria-label']
    except:
        Like = '0' 
    # Angry
    try:
        ANGER = feedback.find('span', {'data-testid':'UFI2TopReactions/tooltip_ANGER'}).find('a').attrs['aria-label']
    except:
        ANGER = '0'
    # HAHA
    try:
        HAHA = feedback.find('span', {'data-testid':'UFI2TopReactions/tooltip_HAHA'}).find('a').attrs['aria-label']
    except:
        HAHA = '0'
    # ç•™è¨€
    try:
        commentcount = feedback.find('a', {'data-testid':'UFI2CommentsCount/root'}).text
    except:
        commentcount = '0' 
    # åˆ†äº«
    try:
        share = feedback.find('span', {'class':'_355t _4vn2'}).text
    except:
        share = '0' 
    return pd.DataFrame(
        data = [{'Name':Name,
                 'ID':ID,
                 'Link':Link,
                 'Time':Time,
                 'Content':Content,
                 'Like':Like,
                 'ANGER':ANGER,
                 "HAHA":HAHA,
                 'commentcount':commentcount,
                 'share':share}],
        columns = ['Name', 'ID', 'Time', 'Content', 'Like', 'ANGER', 'HAHA', 'commentcount', 'share', 'Link'])
```

### æ“·å–ç²‰çµ²ç•™è¨€è³‡è¨Š
é€™é‚Šè¦ç•™æ„é›–ç„¶éƒ½æ˜¯ç²‰çµ²ç•™è¨€ï¼Œä½†å¯¦éš›ä¸Šåˆ†æˆã€Œå›æ‡‰è²¼æ–‡çš„ç•™è¨€ã€èˆ‡ã€Œå›æ‡‰ç•™è¨€çš„ç•™è¨€ã€ã€‚
å‡½æ•¸ä¸­çš„ç¬¬ä¸€å€‹è¿´åœˆæ˜¯ç”¨ä¾†æŠ“ã€Œå›æ‡‰è²¼æ–‡çš„ç•™è¨€ã€ï¼Œç¬¬äºŒå€‹å‰‡æ˜¯æŠ“ã€Œå›æ‡‰ç•™è¨€çš„ç•™è¨€ã€ã€‚è®€è€…å¯ä»¥è‡ªè¡Œæ¯”è¼ƒä¸€ä¸‹å…©å€‹è¿´åœˆä¸­ä¸åŒçš„åœ°æ–¹ã€‚

```python
# ç•™è¨€
def CrawlComment(soup):
    Comments = pd.DataFrame()
    # poæ–‡å€å¡Š
    userContent = soup.find('div', {'class':'_5pcr userContentWrapper'})
    # ç”¨æˆ¶ç•™è¨€å€
    userContent = soup.find('div', {'class':'_5pcr userContentWrapper'})
    # å›æ‡‰è²¼æ–‡çš„ç•™è¨€
    for i in userContent.findAll('div', {'data-testid':'UFI2Comment/root_depth_0'}):
        try:
            CommentContent = i.find('span', {'dir':'ltr'}).text
        except:
            CommentContent = 'Sticker'
        Comment = pd.DataFrame(data = [{'CommentID':i.find('a', {'class':' _3mf5 _3mg0'}).attrs['data-hovercard'].split('id=',2)[1],
                                 'CommentName':i.find('img').attrs['alt'],
                                 'CommentTime':i.find('abbr',{'class':'livetimestamp'}).attrs['data-tooltip-content'],
                                 'CommentContent':CommentContent,
                                 'Link':driver.current_url}],
                        columns = ['CommentID', 'CommentName', 'CommentTime', 'CommentContent', 'Link'])
        Comments = pd.concat([Comments, Comment], ignore_index=True)
    
    # å›æ‡‰ç•™è¨€çš„ç•™è¨€
    for i in userContent.findAll('div', {'data-testid':'UFI2Comment/root_depth_1'}):
        try:
            CommentContent = i.find('span', {'dir':'ltr'}).text
        except:
            CommentContent = 'Sticker'
        Comment = pd.DataFrame(data = [{'CommentID':i.find('a', {'class':' _3mf5 _3mg1'}).attrs['data-hovercard'].split('id=',2)[1],
                                 'CommentName':i.find('img').attrs['alt'],
                                 'CommentTime':i.find('abbr',{'class':'livetimestamp'}).attrs['data-tooltip-content'],
                                 'CommentContent':CommentContent,
                                 'Link':driver.current_url}],
                        columns = ['CommentID', 'CommentName', 'CommentTime', 'CommentContent', 'Link'])
        Comments = pd.concat([Comments, Comment], ignore_index=True)        
    return Comments
```

### å¯¦ä½œ
ä»Šå¤©è¦çˆ¬çš„æ˜¯å°ç£å¤§å“¥å¤§çš„ç²‰çµ²åœ˜é é¢ã€‚æš«æ™‚å…ˆè¨­å®šåŠ è¼‰20æ¬¡è³‡æ–™ï¼Œè‹¥æƒ³æŠ“æ›´å¤š/æ›´å°‘è³‡æ–™çš„è©±å¯ä»¥è‡ªè¡Œèª¿æ•´nçš„æ•¸å€¼ã€‚
æ³¨æ„é€™è£¡æœƒé€éSeleniumé–‹å•Ÿä¸€å€‹Chromeç€è¦½å™¨ï¼Œè‹¥æ²’æœ‰ä¸‹è¼‰çš„äººå¯ä»¥åƒè€ƒåœ¨Windowsä¸Šå®‰è£Python & Selenium + ç°¡æ˜“æ•™å­¸ã€‚

```python
driver = webdriver.Chrome()
Links = FindLinks(url = 'https://facebook.com/taiwanmobile/',
                  n = 20)
Links
```

```python
æ¥è‘—å…ˆå‰µé€ å…©å€‹DataFrameï¼Œä¸€å€‹æ”¾æ–‡ç« å…§å®¹ï¼Œå¦ä¸€å€‹æ”¾ç•™è¨€å…§å®¹ã€‚
æˆ‘è®“ç¨‹å¼è‡ªå‹•è¼¸å‡ºç›®å‰çš„è™•ç†çš„ç¶²å€ï¼Œè‹¥æœ‰ç„¡æ³•æŠ“å‡ºçš„é é¢ä¹Ÿæœƒé€å‡ºä¸€å€‹è¨Šæ¯æé†’ã€‚æ–¹ä¾¿æˆ‘å€‘å¾ŒçºŒè¿½è¹¤å“ªè£¡å‡ºç¾éŒ¯èª¤ã€‚
```

```python
# æŠ“ä¸‹ä¾†æ‰€æœ‰ç•™è¨€
PostsInformation = pd.DataFrame()
PostsComments = pd.DataFrame()
for i in Links:
    print('Dealing with: ' + i)
    try:
        expand(i)
        soup = BeautifulSoup(driver.page_source)
        PostsInformation = pd.concat([PostsInformation, PostContent(soup)],ignore_index=True)
        PostsComments = pd.concat([PostsComments, CrawlComment(soup)],ignore_index=True)
    except:
        print('Load Failed: ' + i)
```
